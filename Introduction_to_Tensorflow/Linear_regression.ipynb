**Splitting the dataset into the training and test set**

Splitting my dataset into an 80-20 split. That is 80% on the training and 20% on testing.

The random_state here is just to ensure that you get the same result when you are running this code in your notebook



```python
y = dataset["Y house price of unit area"]
```

```python
X = dataset[["X1 transaction date" ,"X2 house age", "X3 distance to the nearest MRT station", "X4 number of convenience stores","X5 latitude", "X6 longitude"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

**Training the Simple Linear Regression model on the training set**

It should be noted that we shouldn't introduce our test data to our model while training

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

### **Predicting the test set results**

```python
y_pred = regressor.predict(X_test)
```

### Visualising the test set results

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

We move over to evaluating our model

```python
#Calculating the mean absolute error, mean sqaured error and the root mean squared error
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

Residuals

```python
#We move ahead to exploring the residuals to ensure everything is alright with our code
sns.distplot((y_test-y_pred),bins=50);
```
